```python
# Loading the .csv data
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('heart_2022_no_nans.csv')
df
```

# Formulated Question

1.) What are the top 10 states that has the highest percentage of people that had heart attacks? (Bar Chart)

2.) What is the prevalence of heart attack among males and females? (Pie Chart)

# 1.) What are the top 10 states that has the highest percentage of people that had heart attacks? (Bar Chart)

```python
# Filter data for people who had a heart attack
heart_attack_data = df[df['HadHeartAttack'] == 'Yes']

# Group by 'State' and calculate the percentage of heart attacks
state_heart_attack_percentage = heart_attack_data.groupby('State')['HadHeartAttack'].count() / df.groupby('State')['HadHeartAttack'].count() * 100
state_heart_attack_percentage = state_heart_attack_percentage.sort_values(ascending=False)

# Select the top 10 states
top_10_states = state_heart_attack_percentage[:10]

# Create a bar chart for the top 10 states
plt.figure(figsize=(12, 6))
plt.bar(top_10_states.index, top_10_states.values)
plt.xlabel('State')
plt.ylabel('Percentage of Heart Attacks')
plt.title('Top 10 States with Highest Percentage of People with Heart Attacks')
plt.xticks(rotation=45)
plt.show()
```

For Question 1 - What are the top 10 states that has the highest percentage of people that had heart attacks? (Bar Chart)

Observation:

The bar chart illustrating the top 10 states with the highest percentage of people with heart attacks reveals noticeable geographic variations in heart attack risk. Certain states, like "Arkansas" and "Florida", stand out with considerably higher percentages compared to others in the top 10. While this visualization highlights these leading states, it's important to consider it with the overall prevalence across all states. To gain a deeper understanding between the differences, further investigation is needed to explore potential contributing factors such as demographics and health behaviors specific to each state. Ultimately, the bar chart serves as a valuable starting point for targeted public health interventions and research endeavors aimed at lessening heart attack risk.

# 2.) What is the prevalence of heart attack among males and females? (Pie Chart)

```python
# Group data by sex and count heart attack cases
sex_heart_disease = df[df['HadHeartAttack'] == 'Yes'].groupby('Sex')['HadHeartAttack'].count()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(sex_heart_disease, labels=sex_heart_disease.index, autopct='%1.1f%%', startangle=90)
plt.title('Prevalence of Heart Disease Among Different Sexes')
plt.show()
```

For Question 2 - What is the prevalence of heart attack among males and females? (Pie Chart)

Observation:

The pie chart shows that more males had heart attacks than females. About 63% of the heart attacks happened to males. Women had about 37% of the heart attacks. This means that heart attacks are more common in males. The gap in the percentage may be indicative of how males have a more different lifestyle that females which is why they have a higher number of heart attacks. However, it's important to remember that women can have heart attacks too even though their percentages are lower, so everyone should be aware of the signs.
