# Sample Problem
Given the following sets of data:

x = [0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40]

y = [1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23]

Use linear model and interpret your data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Inputting the given data
x = np.array ([0.50, 1.00, 1.30, 1.60, 2.00, 2.20, 2.50, 3.10, 3.90, 4.40])
y = np.array ([1.90, 1.50, 1.20, 1.00, 0.80, 0.78, 0.65, 0.46, 0.30, 0.23])

# Calculate the slope and y-intercept
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the data points and show the regression line
plt.scatter(x, y, label='Data Points')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Linear Regression', color="red")
plt.legend()
plt.show()

# Print the values
print(f"The value of the slope is {slope}")
print(f"The value of the y-intercept is {intercept}")
```

INTERPRETATION:

The graph illustrates a clear negative correlation between variables x and y. As the values of X increase, the corresponding values of Y tend to decrease, following a general downward trend. This relationship is represented by the regression line, which provides a linear approximation of the data points. The slope of the regression line, calculated to be -0.40. The y-intercept of 1.79 represents the predicted value of y when x is 0. Overall, the graph suggests a moderate linear relationship between the two variables, although some deviation from the regression line indicates that the relationship is not perfectly linear.
