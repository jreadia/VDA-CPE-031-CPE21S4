**2.) Creating a DataFrame from CSV file**

```python
lab_data = pd.read_csv(path)
games_list = pd.DataFrame(lab_data)
games_list
```

**3.) Creating a DataFrame from a dictionary of lists**

```python
# Manually Creating a dictionary based from the original csv
# Extract data for the first 5 observations for 'Name', 'Developer', 'Producer', 'Genre', 'Operating System', and 'Date Released
game_data = {
    'Name': games_list['Name'].tolist()[:5],
    'Developer': games_list['Developer'].tolist()[:5],
    'Producer': games_list['Producer'].tolist()[:5],
    'Genre': games_list['Genre'].tolist()[:5],
    'Operating System': games_list['Operating System'].tolist()[:5],
    'Date Released': games_list['Date Released'].tolist()[:5]
}

# Create the DataFrame
game_df = pd.DataFrame(game_data)

# Print the DataFrame
game_df
```

**4.) Creating a DataFrame from a list of dictionaries**

```python
game_data_list = []

for i in range(5):  # 'For Loop' to get the first 5 observations
    game_data_list.append({
        'Name': games_list.loc[i, 'Name'],
        'Developer': games_list.loc[i, 'Developer'],
        'Producer': games_list.loc[i, 'Producer'],
        'Genre': games_list.loc[i, 'Genre'],
        'Operating System': games_list.loc[i, 'Operating System'],
        'Date Released': games_list.loc[i, 'Date Released']
    })

# Create the DataFrame
games_df_from_list = pd.DataFrame(game_data_list)

# Print the DataFrame
games_df_from_list
```
