# Sample Program

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the data points and the regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.show()
```

# Scatter Plot Diagram using the Given Table

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/CPE 031 - Visualization and Data Analysis/Adia - Exercise 10 - Linear Regression/Linear and Pearson's.csv")
df
```

```python
# Creating a Scatter Plot for Linear Regression (Slope Intercept)

# Extract data
x = df['Proficiency Exam Score (x)'].head(9)
y = df['Course Grade (y)'].head(9)

# Calculate slope and intercept
slope, intercept = np.polyfit(x, y, 1)

# Create regression line
y_pred = slope * x + intercept

# Plot data and regression line
plt.scatter(x, y)  # Scatter plot of original data
plt.plot(x, y_pred, color='green', label='Regression Line')  # Regression line
plt.xlabel('Proficieny Exam Score (x)')
plt.ylabel('Course Grade (y)')
plt.title('Linear Regression', color="green")
plt.legend()
plt.show()

print(f"The slope of the regression line is: {slope}")
print(f"The intercept of the regression line is: {intercept}")
```
