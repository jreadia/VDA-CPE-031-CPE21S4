# Matplotlib Pyplot

```python
import matplotlib.pyplot as plt #pyplot submodule and imported under plt alias
import numpy as np

xpoints = np.array([0,6]) # plotting x and y points
ypoints = np.array([0,250])

plt.plot(xpoints,ypoints) #plot() is used to draw points
plt.show()
```

```python
# Draw a line in a diagram from position (1,3) to position (8,10)
xpoints_2 = np.array([1,8]) # plotting x and y points
ypoints_2 = np.array([3,10])

plt.plot(xpoints_2,ypoints_2) #plot() is used to draw points
plt.show()
```

* Plotting without line




```python
xpoints = np.array([0,6]) # plotting x and y points
ypoints = np.array([0,250])

plt.plot(xpoints,ypoints, linestyle = 'None', marker = 'o') #plot() is used to draw points
plt.show()
```

```python
# Draw a line in a diagram from position (1,3) to (2,8) to (6,1) to (8,10)
xpoints_4 = np.array([1,2,6,8]) # plotting x and y points
ypoints_4 = np.array([3,8,1,10])

plt.plot(xpoints_4,ypoints_4) #plot() is used to draw points
plt.show()
```

```python
#plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker = 'x') #plot() is used to draw points
```

* Format Strings fmt

```python
#plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, 'o:b') #marker | line | color
plt.show()
```

```python
#plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, 'o:b', ms = 20) #marker | line | color | marker size
plt.show()
```

* Marker Color - You can use the argument markeredgecolor or mec for short

```python
#plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, 'o:b', ms = 10, mec = 'r') #marker | line | color | marker size | marker color
plt.show()
```

* Marker Face Color or mfc

```python
#plotting x and y points
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, 'o:b', ms = 10, mec = 'green', mfc = 'red') #marker | line | color | marker size | marker color | marker face color
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
```

* Matplot Labels and Title - You can use the xlabel() and ylabel()

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km", color = 'blue') #labels
plt.ylabel("Speed in km/s")
plt.show()
```

* Matplot Grid - You can use grid() to add grid lines

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km", color = 'blue') #labels
plt.ylabel("Speed in km/s")
plt.grid() #grid
plt.show()

# ---------------------------------- #
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km", color = 'blue') #labels
plt.ylabel("Speed in km/s")
plt.grid(axis = 'y') #grid
plt.show()

# ---------------------------------- #
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km", color = 'blue') #labels
plt.ylabel("Speed in km/s")
plt.grid(axis = 'x', linestyle = '--', linewidth = 0.5) #grid
plt.show()
```

* Show Multiple Plots

```python
#plot 1
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(xpoints,ypoints) #plot() is used to draw points

#plot 2
xpoints = np.array([1,2,6,8])
ypoints = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(xpoints,ypoints)

plt.grid(axis = 'x', linestyle = '--', linewidth = 0.5)
plt.show()
```

* Matplot Scatter - The scatter function plots one dor for each observation

```python

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.scatter(xpoints,ypoints)

xpoints = np.array([1,2,6,8])
ypoints = np.array([10,20,30,40])

plt.scatter(xpoints,ypoints)
plt.show()
```

* Matplotlib Bars - You can use the bar() to draw bars

```python
x = np.array(["Student 1", "Student 2", "Student 3"])
y = np.array([90,85,86])

plt.bar(x,y)
plt.show()
```

* Matplotlib Pie Chart - You can use pie() to draw pie chart

```python
g = np.array([35,25,25,15])
grader = ["Grade 1 = 35%", "Grade 2 = 25%", "Grade 3 = 25%", "Grade 4 = 15%"]
plt.pie(g, labels = grader)
plt.legend(title = 'Grader', loc='upper left', prop={'size': 8})
plt.show()
```
